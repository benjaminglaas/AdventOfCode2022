```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
lines = [x.replace('\n','') for x in lines]
```

```python
x = True
found = False
runner = 0
while x:
    s = set(lines[0][runner:runner+4])
    print(s)
    if len(s) == 4:
        found = True
    if found:
        x = False
        print(lines[runner:runner+4])
        print(runner + 4)
    runner = runner + 1
```
