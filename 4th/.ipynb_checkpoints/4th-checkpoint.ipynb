```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
lines = [x.replace('\n','') for x in lines]
```

```python
lines[0].split(',')[0].split('-')
```

```python
res = 0
for l in lines:
    d1 = l.split(',')[0].split('-')[0]
    d2 = l.split(',')[0].split('-')[1]
    d3 = l.split(',')[1].split('-')[0]
    d4 = l.split(',')[1].split('-')[1]
    
    a = set(range(int(d1), int(d2)+1))
    b = set(range(int(d3), int(d4)+1))
    
    if a.issubset(b) or b.issubset(a):
        print(a)
        print(b)
        res+=1
```

```python
res = 0
for l in lines:
    d1 = l.split(',')[0].split('-')[0]
    d2 = l.split(',')[0].split('-')[1]
    d3 = l.split(',')[1].split('-')[0]
    d4 = l.split(',')[1].split('-')[1]
    
    a = set(range(int(d1), int(d2)+1))
    b = set(range(int(d3), int(d4)+1))
    
    if a.intersection(b):
        print(a)
        print(b)
        res+=1
```
