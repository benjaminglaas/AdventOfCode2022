```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
print(alphabet)
```

```python
lines
```

```python
lines = [x.replace('\n','') for x in lines]
```

```python
lines
```

```python
res = 0
for i in lines:
    l = int(len(i))
    l1 = l -1 
    s = int(l/2)
    s1 = int(s)
    first = i[0:s]
    second = i[s1:l1+1]
    val = set(first).intersection(second)
    if len(val) != 0:
        element = min(val)
        res += alphabet.index(element) + 1
    else:
        print(first + ', ' + second)
        print(l)
        print(i)
```

```python
res = 0
for i in range(int(len(lines)/3)):
    
    first = lines[3*i]
    second = lines[3*i + 1]
    third = lines[3*i + 2]
    val = set(first) & set(second) & set(third)

    
    if len(val) != 0:
        element = min(val)
        res += alphabet.index(element) + 1
    else:
        print(first + ', ' + second + ' , ' + third)
        print(i)
```
